```python
import requests as req
from bs4 import BeautifulSoup as bfs

url='https://www.youtube.com/results?search_query=개발자'

html = req.get(url).text

soup = bfs(html,'html.parser')
```

```python
print(soup)
```

```python
title_list=[]
url_list=[]
for tag in soup.select('li > div > div > div.yt-lockup-content > h3 > a'):
    print(tag.text,'https://www.youtube.com/'+tag['href'])
    title_list.append(tag.text)
    url_list.append('https://www.youtube.com/'+tag['href'])
```
